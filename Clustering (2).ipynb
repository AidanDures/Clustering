```python
def compute_closest_centroids(centroids,inputs):

    #centroids: k x m array 
    #inputs: n x m array 
    #Returns closest_centroids: n-dim vector of indices of closest-centroids

    n,m = inputs.shape
    k=centroids.shape[0]
    closest_centroids=np.zeros(n)
    
    for i in range(n):

        shortest_distance=0
        
        for j in range(k):

            distance = 0

            for l in range(m):

                distance+=(inputs[i][l]-centroids[j][l])**2
            
            if j==0 or distance<shortest_distance:

                shortest_distance=distance

                closest_centroids[i]=j
                
    return closest_centroids
```

```python
def create_centroid(input):

    #input: a x m array

    #ouputs centroid - mean of inputs

    a,m=input.shape

    centroid = np.zeros(m)

    for i in range(a):

        for j in range(m):

            centroid[j]+= (1/a)*input[i][j]
            
    return centroid

#I didn't need this function in the end
```
